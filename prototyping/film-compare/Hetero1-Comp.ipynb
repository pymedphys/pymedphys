```python
monaco_cc_dose = pydicom.read_file(str(MONACOCC_DICOM_DIR.joinpath(f'DoseCHECK1_' + CC_file + f'_Dose.dcm')), force=True)
monaco_mcw_dose = pydicom.read_file(str(MONACOMCW_DICOM_DIR.joinpath(f'DoseCHECK1_' + MCW_file + f'_Dose.dcm')), force=True)
monaco_mcm_dose = pydicom.read_file(str(MONACOMCM_DICOM_DIR.joinpath(f'DoseCHECK1_' + MCM_file + f'_Dose.dcm')), force=True)
dosecheck_dose = pydicom.read_file(str(DOSECHECK_DICOM_DIR.joinpath(f'SDCDoseFile.dcm')), force=True)

dicom_cc_plan = pydicom.read_file(str(MONACOCC_DICOM_DIR.joinpath(f'DoseCHECK1_' + CC_file + f'.dcm')), force=True)
dicom_mcw_plan = pydicom.read_file(str(MONACOMCW_DICOM_DIR.joinpath(f'DoseCHECK1_' + MCW_file + f'.dcm')), force=True)
dicom_mcm_plan = pydicom.read_file(str(MONACOMCM_DICOM_DIR.joinpath(f'DoseCHECK1_' + MCM_file + f'.dcm')), force=True)
```
